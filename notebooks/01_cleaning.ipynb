```python
import pandas as pd
import datetime as dt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../data/online_retail_II.csv')
print(df.shape)
df.head()
```

```python
print(df.dtypes)
print("\nNull values:\n", df.isnull().sum())
```

```python
# Drop rows with no customer
df = df.dropna(subset=['Customer ID'])

# Remove cancellations (invoices starting with C)
df = df[~df['Invoice'].astype(str).str.startswith('C')]

# Remove bad quantities and prices
df = df[df['Quantity'] > 0]
df = df[df['Price'] > 0]

# Fix data types
df['Customer ID'] = df['Customer ID'].astype(int)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

print("Clean dataset shape:", df.shape)
```

```python
# Revenue per transaction
df['TotalPrice'] = df['Quantity'] * df['Price']

# Date parts for Tableau later
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['MonthYear'] = df['InvoiceDate'].dt.to_period('M').astype(str)

df.head()
```

```python
df.to_csv('../data/online_retail_clean.csv', index=False)
print("✅ Saved to data/online_retail_clean.csv")
```
